```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("healthcare.csv")
```

```python
data.head()
```

```python
chest_pain = data[data['DRG Definition'] == '313 - CHEST PAIN']
```

```python
chest_pain
```

```python
alabama_chest_pain = chest_pain[chest_pain['Provider State'] == "AL"]
```

```python
alabama_chest_pain
```

```python
costs = alabama_chest_pain[' Average Covered Charges '].values
```

```python
costs
```

```python
plt.boxplot(costs)
plt.title("Boxplot of Chest Pain Charges in Alabama")
plt.ylabel("Charges ($)")
plt.show()
```

```python
states = chest_pain['Provider State'].unique()
datasets = []

for state in states:
    state_data = chest_pain[chest_pain['Provider State'] == state][' Average Covered Charges '].values
    datasets.append(state_data)
```

```python
datasets
```

```python
plt.figure(figsize=(20, 6))

plt.boxplot(datasets, labels=states)
plt.title("Chest Pain Charges by State")
plt.ylabel("Charges ($) In Dollats")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
