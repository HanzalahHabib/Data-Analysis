## Function to remove extra spaces and put underscore (_) instead of them

```python
def spaces_to_underscore(dataframe):
    dataframe.columns = dataframe.columns.str.replace(" ","_")
    return dataframe
```

## Creating Bins

```python
bins = (9.0,50000,100000,150000,200000,250000,300000,350000,400000,450000,5000000)
```

## Cutting dataset using bins

```python
cuts = pd.cut(dataset.Total_cases,bins)
```

```python
cuts
```

## Number of Countries falling in Each Bin

```python
frequency = pd.value_counts(cuts)
frequency
```

## Setting index for frequency

## Importing Matplotlib library for plotting bars and graphs

```python
import matplotlib.pyplot as plt
```

## Plotting a bar graph showing number of Countries falling in which bin according to the total covid cases reported
