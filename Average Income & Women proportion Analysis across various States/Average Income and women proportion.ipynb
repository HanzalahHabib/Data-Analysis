```python
s0["Hispanic"] = s0.Hispanic.str.replace("%","")
```

```python
s0["Hispanic"] = s0["Hispanic"].astype(float).round().astype(int) * 100
```

```python
s0
```

```python
import numpy as np
```

```python
percent_cols = [col for col in s0.columns if s0[col].astype(str).str.contains('%').any()]

# Apply the transformation safely
for col in percent_cols:
    s0[col] = (
        s0[col]
        .astype(str)
        .str.rstrip('%')
        .replace(["NaN", "inf", "None", "nan"], np.nan)  # Handle bad values
        .astype(float)
        .round()
        .astype("Int64") * 100
    )
```

```python
s0
```

```python
s0['Income'] = s0['Income'].astype(str).str.lstrip('$')
```

```python
s0
```

```python
s0['Income'] = (
    s0['Income']
    .astype(str)  # Ensure everything is a string
    .str.replace(r"[^\d.]", "", regex=True)  # Keep numbers and dots only
    .astype(float)  # Convert to float first # Finally convert to integer
)

print(s0['Income'])
```

```python
s0
```

```python
s0[["Males", "Females"]] = s0["GenderPop"].str.split("_", expand=True)

s0["Males"] = s0["Males"].str.rstrip("M").astype(int)
s0["Females"] = s0["Females"].str.rstrip("F").astype(int)
```

```python
s0
```

```python
s0.drop("GenderPop",axis=1,inplace=True)
```

```python
s0.drop("Unnamed: 0",axis=1,inplace=True)
```

# The first visualization your boss wants you to make is a scatterplot that shows average income in a state vs proportion of women in that state.

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x = s0.Income,y=s0.Females)
plt.ylabel("Female")
plt.xlabel("Income")
plt.title("A scatterplot that shows average income in a state vs proportion of women in that state")
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(s0['Females'], s0['Income'], color='purple', alpha=0.8)
plt.title('Average Income vs Proportion of Women per State')
plt.xlabel('Proportion of Women')
plt.ylabel('Average Income ($)')
plt.grid(True)

plt.show()
```
